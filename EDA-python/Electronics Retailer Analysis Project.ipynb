```python
# Convert the 'Birthday' column to datetime format
customers_df['Birthday'] = pd.to_datetime(customers_df['Birthday'])

# Get the current system date
current_date = pd.to_datetime(datetime.now().date())

# Create a new column 'Age_Difference' with the difference between the current date and birthday
customers_df['Age_Difference'] = (current_date - customers_df['Birthday']).dt.days // 365  # Difference in years
```

```python
# Define bins and labels for the age groups
bins = [0, 18, 30, 45, 60, 100]  # Age ranges
labels = ['0-18', '19-30', '31-45', '46-60', '61+']  # Labels for the age ranges

# Create a new column 'Age_Group' with the age bins
customers_df['Age_Group'] = pd.cut(customers_df['Age_Difference'], bins=bins, labels=labels, right=False)

customers_df
```

```python
# Count the occurrences of each age group
age_group_counts = customers_df['Age_Group'].value_counts()

# Plot the bar chart
age_group_counts.plot(kind='bar', color='skyblue', figsize=(8, 6))

# Add titles and labels
plt.title('Distribution of Age Groups', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Number of People', fontsize=14)

# Display the chart
plt.show()
```

### Summary
#### The age group with the highest number of people is 61+, followed by 31-45 and 46-60. The age groups 19-30 and 0-18 have significantly lower numbers of people.

```python
# Clean 'Unit Cost USD' and 'Unit Price USD' columns by removing $, spaces, and commas
product_df['Unit Cost USD'] = product_df['Unit Cost USD'].replace({'\$': '', ' ': '', ',': ''}, regex=True).astype(float)
product_df['Unit Price USD'] = product_df['Unit Price USD'].replace({'\$': '', ' ': '', ',': ''}, regex=True).astype(float)

# Define min and max for cost and price
Min_Cost = product_df['Unit Cost USD'].min()
Max_Cost = product_df['Unit Cost USD'].max()
Min_Price = product_df['Unit Price USD'].min()
Max_Price = product_df['Unit Price USD'].max()

# Create bins for cost and price
Cost_bins = np.linspace(Min_Cost, Max_Cost, 10)  # 10 cost ranges
Price_bins = np.linspace(Min_Price, Max_Price, 10)  # 10 price ranges

# Create the cost groups using the bins
product_df['Cost_groups'] = pd.cut(product_df['Unit Cost USD'], bins=Cost_bins, right=False)
product_df['Price_groups'] = pd.cut(product_df['Unit Price USD'], bins=Price_bins, right=False)

# Display the DataFrame
product_df
```

```python
product_df['Cost_groups'].value_counts(), product_df['Price_groups'].value_counts()
```

```python
# Set up the figure and axes for two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plotting Horizontal Histogram for Unit Cost Distribution
product_df['Cost_groups'].value_counts(ascending=True).plot(kind='barh', color='skyblue', ax=ax1)
ax1.set_title('Distribution of Unit Cost USD', fontsize=16)
ax1.set_xlabel('Number of Products', fontsize=14)
ax1.set_ylabel('Unit Cost USD', fontsize=14)

# Plotting Horizontal Histogram for Unit Price Distribution
product_df['Price_groups'].value_counts(ascending=True).plot(kind='barh', color='lightgreen', ax=ax2)
ax2.set_title('Distribution of Unit Price USD', fontsize=16)
ax2.set_xlabel('Number of Products', fontsize=14)
ax2.set_ylabel('Unit Price USD', fontsize=14)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Function to detect outliers using IQR

def detect_outliers_iqr(data, column_name):
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column_name] < lower_bound) | (data[column_name] > upper_bound)]
    return outliers

# Detect outliers in 'Unit Cost USD' and 'Unit Price USD'
unit_cost_outliers = detect_outliers_iqr(product_df, 'Unit Cost USD')
unit_price_outliers = detect_outliers_iqr(product_df, 'Unit Price USD')

print("Outliers in 'Unit Cost USD':")
print(unit_cost_outliers)
print("\
Outliers in 'Unit Price USD':")
print(unit_price_outliers)
```

```python
# Create box plots for Unit Cost USD and Unit Price USD
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.boxplot(x=product_df['Unit Cost USD'], ax=ax1)
ax1.set_title('Distribution of Unit Cost USD')
ax1.set_xlabel('Unit Cost USD')

sns.boxplot(x=product_df['Unit Price USD'], ax=ax2)
ax2.set_title('Distribution of Unit Price USD')
ax2.set_xlabel('Unit Price USD')

plt.tight_layout()
plt.show()
```

### The outliers in 'Unit Cost USD' are primarily high-end products like refrigerators and televisions, which have significantly higher costs compared to other products in the dataset. These products represent about 7.27% of the total products, indicating that they are a small but distinct group with higher costs.
